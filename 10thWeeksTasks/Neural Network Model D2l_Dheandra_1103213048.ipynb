```python
import numpy as np
import pandas as pd
import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader
from d2l import torch as d2l
```

```python
# Load dataset 1
dataset_path_1 = "/content/drive/MyDrive/dataset/housepricedata.csv"
data_1 = pd.read_csv(dataset_path_1)

# Load dataset 2
dataset_path_2 = "/content/drive/MyDrive/dataset/train.csv"
data_2 = pd.read_csv(dataset_path_2)
```

```python
# Preprocess data
# Assume the last column is the target
features = data_1.iloc[:, :-1].values
labels = data_1.iloc[:, -1].values
```

```python
scaler_X = StandardScaler()
features = scaler_X.fit_transform(features)

scaler_y = StandardScaler()
labels = scaler_y.fit_transform(labels.reshape(-1, 1))

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
# Define dataset
class HousePriceDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_dataset = HousePriceDataset(X_train, y_train)
test_dataset = HousePriceDataset(X_test, y_test)
```

```python
# Define model
class MLP(nn.Module):
    def __init__(self, input_dim):
        super(MLP, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
        )

    def forward(self, X):
        return self.net(X)
```

```python
# Initialize model and loss
input_dim = X_train.shape[1]
model = MLP(input_dim)
loss_fn = nn.MSELoss()
```

```python
# Training loop
def train(model, train_loader, loss_fn, optimizer, num_epochs):
    for epoch in range(num_epochs):
        model.train()
        for X_batch, y_batch in train_loader:
            optimizer.zero_grad()
            y_pred = model(X_batch)
            loss = loss_fn(y_pred, y_batch)
            loss.backward()
            optimizer.step()
```

```python
# Training
batch_size = 64
num_epochs = 100
learning_rate = 0.001
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

train(model, train_loader, loss_fn, optimizer, num_epochs)
```

```python
# Evaluation
model.eval()
with torch.no_grad():
    y_pred_train = model(torch.tensor(X_train, dtype=torch.float32))
    y_pred_test = model(torch.tensor(X_test, dtype=torch.float32))

train_loss = loss_fn(y_pred_train, torch.tensor(y_train, dtype=torch.float32)).item()
test_loss = loss_fn(y_pred_test, torch.tensor(y_test, dtype=torch.float32)).item()

print(f"Train Loss: {train_loss}, Test Loss: {test_loss}")
```
