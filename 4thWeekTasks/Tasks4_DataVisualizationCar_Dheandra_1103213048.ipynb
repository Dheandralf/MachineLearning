```python
# Mengimpor library pandas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Menghubungkan Colab dengan Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Path ke file bikes.csv di Google Drive
file_path = '/content/drive/My Drive/dataset/car.csv'

# Baca file CSV menggunakan pandas
df = pd.read_csv(file_path)
```

```python
# Filter hanya fitur numerik
numeric_features = df.select_dtypes(include=['int64', 'float64'])

# Tampilkan boxplot dari fitur numerik
plt.figure(figsize=(10, 6))
numeric_features.boxplot()
plt.title('Boxplot dari Fitur Numerik')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter hanya fitur numerik
numeric_features = df.select_dtypes(include=['int64', 'float64'])

sns.pairplot(numeric_features)
plt.title('Pairplot dari Fitur Numerik')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for column in numeric_features.columns:
    sns.histplot(df[column], kde=True, bins=20, label=column)
plt.title('Distribution Plot dari Setiap Fitur Numerik')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

```python
# Tampilkan violin plot dari setiap fitur numerik
plt.figure(figsize=(12, 8))
for column in numeric_features.columns:
    sns.violinplot(x=df[column], inner='quartile')
plt.title('Violin Plot dari Setiap Fitur Numerik')
plt.xlabel('Fitur')
plt.ylabel('Nilai')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Hitung korelasi antar fitur numerik
correlation_matrix = numeric_features.corr()

# Tampilkan heatmap dari korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap Korelasi antar Fitur Numerik')
plt.show()
```
