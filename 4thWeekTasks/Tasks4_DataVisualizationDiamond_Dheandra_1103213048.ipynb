```python
# Membuat boxplot untuk setiap fitur numerik
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=diamonds_data[feature])
    plt.title(f'Boxplot dari {feature}')
    plt.xlabel('')
    plt.ylabel('Nilai')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(diamonds_data.select_dtypes(include=['float64', 'int64']))
plt.suptitle("Scatter Plot dari Fitur Numerik dalam Dataset Diamonds", y=1.02)
plt.show()
```

```python
# Memilih fitur numerik
numeric_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat distribution plot untuk setiap fitur numerik
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(3, 3, i)
    sns.histplot(diamonds_data[feature], kde=True)
    plt.title(f'Distribution Plot dari {feature}')
    plt.xlabel('Nilai')
    plt.ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```

```python
# Memilih fitur numerik
numeric_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat violin plot untuk setiap fitur numerik
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(3, 3, i)
    sns.violinplot(y=diamonds_data[feature])
    plt.title(f'Violin Plot dari {feature}')
    plt.ylabel('Nilai')

plt.tight_layout()
plt.show()
```

```python
# Menghitung korelasi antara fitur numerik
correlation_matrix = diamonds_data.corr()

# Membuat heatmap dari korelasi antara fitur numerik
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi antara Fitur Numerik')
plt.show()
```
