```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_path = '/content/drive/My Drive/dataset/IT Salary Survey EU  2020.csv'
df = pd.read_csv (df_path)
df.head()
```

Visualisasi distribusi gaji

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Yearly brutto salary (without bonus and stocks) in EUR'], bins=30, kde=True, color='skyblue')
plt.title('Distribusi Gaji')
plt.xlabel('Gaji (EUR)')
plt.ylabel('Frekuensi')
plt.show()
```

Visualisasi hubungan antara gaji dan pengalaman kerja

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total years of experience', y='Yearly brutto salary (without bonus and stocks) in EUR', data=df, color='green')
plt.title('Hubungan antara Gaji dan Pengalaman Kerja')
plt.xlabel('Total Tahun Pengalaman')
plt.ylabel('Gaji (EUR)')
plt.show()
```

Visualisasi hubungan antara gaji dan tingkat pendidikan

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Seniority level', y='Yearly brutto salary (without bonus and stocks) in EUR')  # Mengganti 'Education' dengan nama kolom yang sesuai
plt.title('Hubungan antara Tingkat Pendidikan dan Gaji')
plt.xlabel('Tingkat Pendidikan')
plt.ylabel('Gaji')
plt.show()
```
